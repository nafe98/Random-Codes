# Mean Accuracy

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Configure global font settings for consistent visual presentation
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 16
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14

# Data representing mean Accuracy for various models across different feature sets
data = {
    'Feature': ['Feature 5', 'Feature 8', 'Feature 16', 'Feature 22', 'All Features'],
    'XGBoost': [0.72, 0.82, 0.92, 0.95, 0.96],
    'Decision Tree': [0.74, 0.84, 0.90, 0.91, 0.91],
    'Random Forest': [0.83, 0.93, 0.98, 0.99, 0.99],
    'Extra Trees': [0.83, 0.94, 0.99, 0.99, 0.99],
    'KNN': [0.84, 0.95, 0.99, 1.00, 1.00]
}

# Convert dictionary into pandas DataFrame
df = pd.DataFrame(data)

# Reshape DataFrame to long format for Seaborn compatibility
df_melted = df.melt(id_vars='Feature', var_name='Model', value_name='Mean Accuracy')

# Configure Seaborn aesthetics for research-level visuals
sns.set(style="whitegrid", context="talk")  # White grid, 'talk' context for larger font sizes

# Use the 'viridis' color palette for distinct colors
palette = sns.color_palette("viridis", n_colors=len(df.columns) - 1)

# Define different line styles and markers for more sophistication
line_styles = ['-', '--', '-.', ':', '-', '--', '-.', ':', '-']  # Different line styles
markers = ['o', 's', 'D', 'v', '^', 'X', 'P', '*', 'p']  # Different markers

# Create figure and axis
plt.figure(figsize=(12, 8))

# Create line plot for mean Accuracy comparison
for i, model in enumerate(df.columns[1:]):
    sns.lineplot(
        x='Feature', 
        y='Mean Accuracy', 
        data=df_melted[df_melted['Model'] == model], 
        label=model, 
        marker=markers[i], 
        linewidth=2.5, 
        linestyle=line_styles[i],
        color=palette[i]  # Use colors from the 'viridis' palette
    )

# Add plot title and axis labels
plt.title('Mean Accuracy of Models Across Feature Sets', fontsize=22, fontweight='bold', pad=20)
plt.xlabel('Feature Set', fontsize=18, labelpad=15)
plt.ylabel('Mean Accuracy', fontsize=18, labelpad=15)

# Enhance grid for better readability
plt.grid(True, which='major', linestyle='--', linewidth=0.8)
plt.grid(True, which='minor', linestyle=':', linewidth=0.5)

# Move the legend to the bottom of the plot with a box style
plt.legend(title='Model', loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3, fontsize=14, frameon=True, shadow=True)

# Adjust layout to prevent overlap and improve appearance
plt.tight_layout()

# Add extra space between the plot and the legend to avoid overlap
plt.subplots_adjust(bottom=0.25)

# Save as a high-quality PNG for research publication
plt.savefig('mean_accuracy_comparison_viridis.png', format='png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Data representing accuracy scores for various models (filtered to include only Machine Learning models)
data = {
    'Model': ['XGB', 'DT', 'RF', 'ET', 'KNN'],
    'Accuracy': [0.92, 0.90, 0.98, 0.99, 0.99],
    'Category': ['Machine Learning'] * 5
}

# Creating a pandas DataFrame from the data
df = pd.DataFrame(data)

# Setting the plot style for a clean, professional look
sns.set(style="whitegrid")

# Creating a bar plot for model accuracy
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Model', y='Accuracy', data=df, hue='Category', palette='Blues')

# Adding plot title and axis labels in Times New Roman font
plt.title('Comparison of Accuracy for Machine Learning Models', fontsize=16, fontname='Times New Roman')
plt.xlabel('Model', fontsize=14, fontname='Times New Roman', labelpad=10)
plt.ylabel('Accuracy', fontsize=14, fontname='Times New Roman', labelpad=10)

# Removing minor ticks and customizing major grid for improved clarity
plt.grid(True, which='major', axis='y', linestyle='--', linewidth=0.7, color='gray')
plt.minorticks_off()

# Setting x and y axis tick font sizes
plt.xticks(fontsize=12, fontname='Times New Roman')
plt.yticks(fontsize=12, fontname='Times New Roman')

# Adding values on top of each bar
for p in ax.patches:
    height = p.get_height()
    if height > 0:  # Check if the bar height is greater than zero
        ax.annotate(f'{height:.2f}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='center', 
                    xytext=(0, 8), 
                    textcoords='offset points', 
                    fontsize=12, fontname='Times New Roman')

# Adjusting layout to prevent label overlap
plt.tight_layout()

# Moving the legend to the bottom with more space to avoid overlap
plt.legend(title='Category', loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=2)

# Saving the figure in a lossless PNG format (best for research journals)
plt.savefig('machine_learning_accuracy_barplot_journal.png', format='png', dpi=300, bbox_inches='tight')

# Displaying the plot
plt.show()
```
