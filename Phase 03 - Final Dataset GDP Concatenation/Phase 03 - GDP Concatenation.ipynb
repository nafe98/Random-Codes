```python
import pandas as pd
```

```python
# Load the provided files
auto_prod_file = 'Exogenus - Auto Production.xlsx'
gdp_file = 'GDP.xlsx'
```

```python
auto_prod_df = pd.read_excel(auto_prod_file)
gdp_df = pd.read_excel(gdp_file)
```

```python
# Convert the "Observation Date" columns to datetime for both datasets
auto_prod_df['Observation Date'] = pd.to_datetime(auto_prod_df['Observation Date'])
gdp_df['Observation Date'] = pd.to_datetime(gdp_df['Observation Date'])
```

```python
# Create a new column that repeats the GDP value for each month within the same quarter
gdp_expanded = gdp_df.set_index('Observation Date').reindex(auto_prod_df['Observation Date'], method='ffill')['GDP'].values

# Add the GDP values to the Auto Production Data
auto_prod_df['GDP'] = gdp_expanded

# Save the resulting dataframe to a new Excel file if needed
auto_prod_df.to_excel("Merged_Auto_Prod_GDP_Data.xlsx", index=False)
```
