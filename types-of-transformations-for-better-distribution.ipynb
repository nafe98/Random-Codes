<p style="font-family: Arials; font-size: 26px; font-style: bold; font-weight: bold; text-align:center; letter-spacing: 3px; color: #5B2C6F ">3.1 LOG TRANSFORMATION:</p>

<p style="font-family:newtimeroman; font-size:200%; text-align:center;"><b>What is Log Transformation?</b></p>

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">Log transformation is a data transformation method in which in which each variable of x will be replaced by log(x) with base 10, base 2, or natural log.. The choice of the logarithm base is usually left up to the analyst and it would depend on the purposes of statistical modeling
When our original continuous data do not follow the bell curve, we can log transform this data to make it as ‚Äúnormal‚Äù as possible so that the statistical analysis results from this data become more valid. In other words, the log transformation reduces or removes the skewness of our original data. The important caveat here is that the original data has to follow or approximately follow a log-normal distribution. Otherwise, the log transformation won‚Äôt work.</p>




```python
sns.displot(df["chol"],kde=True)
plt.title("DISTRIBUTION OF CHOLESTROL LEVEL",fontsize=20)
skewness=str(df["chol"].skew())
kurtosis=str(df["chol"].kurt())
plt.legend([skewness,kurtosis],title=("skewness and kurtosis"))
plt.show()
```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">As we can see the above plot is not normal distribution as there is some positive skewed data</p>

```python
plt.figure(figsize=(20,8))

## log transformation
plt.subplot(1,4,1)
sns.distplot(df["chol"])
plt.title("DISTRIBUTION BEFORE",)
skewness=str(df["chol"].skew())
plt.legend([skewness],title=("skewness"))


plt.subplot(1,4,2)
stats.probplot(df["chol"],dist="norm",plot=plt)
plt.grid()

plt.subplot(1,4,3)
log_target=np.log1p(df["chol"])
ax=sns.distplot(log_target)
plt.title("DISTRIBUTION AFTER LOG ",)
skewness=str(log_target.skew())
plt.legend([skewness],title=("skewness"))
#plt.show()

plt.subplot(1,4,4)
stats.probplot(log_target,dist="norm",plot=plt)
plt.grid()


```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">The above plot is Comparison of Distribution Before Log Transformation and after Log transformation.As we can see the Skewness got Reduced in Transformed Data. The best skewness value is near to zero</p>


<p id="part6"></p>
<p style="font-family: Arials; font-size: 26px; font-style: bold; font-weight: bold; text-align:center; letter-spacing: 3px; color: #5B2C6F ">3.2 SQUARE ROOT TRANSFORMATION:</p>

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">This transformation will give a moderate effect on distribution. The main advantage of square root transformation is, it can be applied to zero values.
Here the x will replace by the square root(x). It is weaker than the Log Transformation.</p>

```python
sns.displot(df["chol"],kde=True)
plt.yscale("log")
plt.title("DISTRIBUTION OF DAILY VACCINATIONS",fontsize=20)
skewness=str(df["chol"].skew())
kurtosis=str(df["chol"].kurt())
plt.legend([skewness,kurtosis],title=("skewness and kurtosis"))
plt.show()
```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">As we can see the above plot is not normal distribution It is  positive skewed data</p>

```python
plt.figure(figsize=(20,8))


plt.subplot(1,4,1)
sns.distplot(df["chol"])
plt.title("DISTRIBUTION BEFORE",)
skewness=str(df["chol"].skew())
plt.legend([skewness],title=("skewness"))


plt.subplot(1,4,2)
stats.probplot(df["chol"],dist="norm",plot=plt)
plt.grid()

plt.subplot(1,4,3)
sqrt_target = df["chol"]**(1/2)
ax=sns.distplot(sqrt_target)
plt.title("DISTRIBUTION AFTER SQRT ",)
skewness=str(sqrt_target.skew())
plt.legend([skewness],title=("skewness"))
#plt.show()

plt.subplot(1,4,4)
stats.probplot(sqrt_target,dist="norm",plot=plt)
plt.grid()
```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">In the above plot we have done Square root Transformtion.As we can see most of the skewness got reduced.It was not best but still it gives us better result </p>

<p id="part7"></p>
<p style="font-family: Arials; font-size: 26px; font-style: bold; font-weight: bold; text-align:center; letter-spacing: 3px; color: #5B2C6F ">3.3 RESIPROCAL TRANSFORMATION:</p>

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">In this transformation, x will replace by the inverse of x (1/x).
The reciprocal transformation will give little effect on the shape of the distribution. This transformation can be only used for non-zero values.</p>

```python
sns.displot(df["trestbps"],kde=True)
plt.title("DISTRIBUTION OF BLOOD PRESSURE",fontsize=20)
skewness=str(df["trestbps"].skew())
kurtosis=str(df["trestbps"].kurt())
plt.legend([skewness,kurtosis],title=("skewness and kurtosis"))
plt.show()
```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">As we can see the above plot is little right skewed. let's apply the resiprocal transformation and see the result</p>

```python
plt.figure(figsize=(20,8))


plt.subplot(1,4,1)
sns.distplot(df["trestbps"])
plt.title("DISTRIBUTION BEFORE",)
skewness=str(df["trestbps"].skew())
plt.legend([skewness],title=("skewness"))


plt.subplot(1,4,2)
stats.probplot(df["trestbps"],dist="norm",plot=plt)
plt.grid()

plt.subplot(1,4,3)
reciprocal_target = 1/df["trestbps"]
sns.distplot(reciprocal_target)
plt.title("DISTRIBUTION AFTER RESIPROCAL ",)
skewness=str(reciprocal_target.skew())
plt.legend([skewness],title=("skewness"))
#plt.show()

plt.subplot(1,4,4)
stats.probplot(reciprocal_target,dist="norm",plot=plt)
plt.grid()
```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">In the above plot There is Distribution of before applying Resiprocal transformation and after applying Resiprocal Transformation. As we can see the Skewness got reduced. </p>

<p id="part8"></p>
<p style="font-family: Arials; font-size: 26px; font-style: bold; font-weight: bold; text-align:center; letter-spacing: 3px; color: #5B2C6F ">3.4 BOX-COX TRANSFORMATION:</p>

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">
Box-cox transformation works pretty well for many data natures. The below image is the mathematical formula for Box-cox transformation.</p>


<img src= "https://miro.medium.com/max/884/1*_nHvGg-OQYtNSSZph-424g.png" width="100%"  align="center"  hspace="5%" vspace="5%"/>

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">All the values of lambda vary from -5 to 5 are considered and the best value for the data is selected. The ‚ÄúBest‚Äù value is one that results in the best skewness of the distribution. Log transformation will take place when we have lambda is zero.</p>






```python
sns.displot(df["thalach"])
plt.title("DISTRIBUTION OF HEART BEAT",fontsize=20)
skewness=str(df["thalach"].skew())
kurtosis=str(df["thalach"].kurt())
plt.legend([skewness,kurtosis],title=("skewness and kurtosis"))
plt.show()
```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">The above plot is the dstribution of Heart Beat as we can see it is little left skewed. so let's apply box cox Transformation and check the results.</p>

```python
plt.figure(figsize=(20,8))


plt.subplot(1,4,1)
sns.distplot(df["trestbps"])
plt.title("DISTRIBUTION BEFORE",)
skewness=str(df["trestbps"].skew())
plt.legend([skewness],title=("skewness"))


plt.subplot(1,4,2)
stats.probplot(df["trestbps"],dist="norm",plot=plt)
plt.grid()

plt.subplot(1,4,3)
bcx_target, lam = boxcox(df["trestbps"])
#lam is the best lambda for the distribution
sns.distplot(bcx_target)
plt.title("DISTRIBUTION AFTER BOX-COX ",)
series=pd.Series(bcx_target)
skewness=str(series.skew())
plt.legend([skewness],title=("skewness"))
#plt.show()

plt.subplot(1,4,4)
stats.probplot(series,dist="norm",plot=plt)
plt.grid()
```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">In the above plot there is distribution of before and after applying box cox transformation .As we can see the box shows very good result.That's great.</p>

<p id="part9"></p>
<p style="font-family: Arials; font-size: 26px; font-style: bold; font-weight: bold; text-align:center; letter-spacing: 3px; color: #5B2C6F ">3.5  YEO-JOHNSON TRANSFORMATION:</p>

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">This is one of the older transformation technique which is very similar to Box-cox transformation but does not require the values to be strictly positive.
This transformation is also having the ability to make the distribution more symmetric.</p>



```python
sns.displot(df["chol"])
plt.yscale("log")
plt.title("DISTRIBUTION OF TOTAL VACINATIONS",fontsize=20)
skewness=str(df["chol"].skew())
plt.legend([skewness],title=("skewness"))
plt.show()
```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">From above plot we can see that there is some +ve skewness in the distribution . let's apply Yeo-Jhonson transformation and see the results also one more thing there is also 0 values present in the data .</p>

```python
plt.figure(figsize=(20,8))


plt.subplot(1,4,1)
sns.distplot(df["chol"])
plt.title("DISTRIBUTION BEFORE",)
skewness=str(df["chol"].skew())
plt.legend([skewness],title=("skewness"))


plt.subplot(1,4,2)
stats.probplot(df["chol"],dist="norm",plot=plt)
plt.grid()

plt.subplot(1,4,3)
yf_target, lam = yeojohnson(df["chol"])
sns.distplot(yf_target)
plt.title("DISTRIBUTION AFTER YEO JHONSON ",)
series=pd.Series(yf_target)
skewness=str(series.skew())
plt.legend([skewness],title=("skewness"))
#plt.show()

plt.subplot(1,4,4)
stats.probplot(yf_target,dist="norm",plot=plt)
plt.grid()
```

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">WOW Skewness got 0.</p>

<p style="font-family: Arials, sans-serif; font-size: 20px; text-align:center; color: rgba(0,0,0,.7)">As we know we have done multiple Transformations Now in this block we will do Transformation on a single Distribution and check which Transformation will give us better Result.</p>

```python
sns.displot(df["chol"])
```

```python
plt.figure(figsize=(25,38))

## log transformation
plt.subplot(5,4,1)
sns.distplot(df["chol"])
plt.title("DISTRIBUTION BEFORE",fontsize=20)
skewness=str(df["chol"].skew())
plt.legend([skewness],title="skewness",fontsize=20)
#plt.show()

plt.subplot(5,4,2)
stats.probplot(df["chol"],dist="norm",plot=plt)
plt.grid()

plt.subplot(5,4,3)
log_target=np.log1p(df["chol"])
sns.distplot(log_target)
plt.title("DISTRIBUTION AFTER LOG ",fontsize=20)
skewness=str(log_target.skew())
plt.legend([skewness],title=("skewness"),fontsize=20)
#plt.show()

plt.subplot(5,4,4)
stats.probplot(log_target,dist="norm",plot=plt)
plt.grid()


#2
### square root transformation
plt.subplot(5,4,5)
sns.distplot(df["chol"])
plt.title("DISTRIBUTION BEFORE",fontsize=20)
skewness=str(df["chol"].skew())
plt.legend([skewness],title="skewness",fontsize=20)
#plt.show()

plt.subplot(5,4,6)
stats.probplot(df["chol"],dist="norm",plot=plt)
plt.grid()

plt.subplot(5,4,7)
sqrt_target = df["chol"]**(1/2)
sns.distplot(sqrt_target)
plt.title("DISTRIBUTION AFTER SQRT ",fontsize=20)
skewness=str(sqrt_target.skew())
plt.legend([skewness],title=("skewness"),fontsize=20)
#plt.show()

plt.subplot(5,4,8)
stats.probplot(sqrt_target,dist="norm",plot=plt)
plt.grid()


#3
## reciprocal transformation
plt.subplot(5,4,9)
sns.distplot(df["chol"])
plt.title("DISTRIBUTION BEFORE",fontsize=20)
skewness=str(df["chol"].skew())
plt.legend([skewness],title=("skewness"),fontsize=20)
#plt.show()

plt.subplot(5,4,10)
stats.probplot(df["chol"],dist="norm",plot=plt)
plt.grid()

plt.subplot(5,4,11)
reciprocal_target = 1/df["chol"]
sns.distplot(reciprocal_target)
plt.title("DISTRIBUTION AFTER RECIPROCAL ",fontsize=20)
skewness=str(reciprocal_target.skew())
plt.legend([skewness],title="skewness",fontsize=20)
#plt.show()

plt.subplot(5,4,12)
stats.probplot(reciprocal_target,dist="norm",plot=plt)
plt.grid()

#4
## 4. Box-Cox Transformation:
plt.subplot(5,4,13)
sns.distplot(df["chol"])
plt.title("DISTRIBUTION BEFORE",fontsize=20)
skewness=str(df["chol"].skew())
plt.legend([skewness],title=("skewness"),fontsize=20)
#plt.show()

plt.subplot(5,4,14)
stats.probplot(df["chol"],dist="norm",plot=plt)
plt.grid()

plt.subplot(5,4,15)
bcx_target, lam = boxcox(df["chol"])
#lam is the best lambda for the distribution

sns.distplot(bcx_target)
plt.title("DISTRIBUTION AFTER BOXCOX ",fontsize=20)
series=pd.Series(bcx_target)
skewness=str(series.skew())
plt.legend([skewness],title=("skewness"),fontsize=20)
#plt.show()

plt.subplot(5,4,16)
stats.probplot(bcx_target,dist="norm",plot=plt)
plt.grid()

# 5. Yeo-Johnson Transformation:
plt.subplot(5,4,17)
sns.distplot(df["chol"])
plt.title("DISTRIBUTION BEFORE",fontsize=20)
skewness=str(df["chol"].skew())
plt.legend([skewness],title=("skewness"),fontsize=20)
#plt.show()

plt.subplot(5,4,18)
stats.probplot(df["chol"],dist="norm",plot=plt)
plt.grid()

plt.subplot(5,4,19)
yf_target, lam = yeojohnson(df["chol"])
sns.distplot(yf_target)
plt.title("DISTRIBUTION AFTER YEOJOHNSON ",fontsize=20)
series=pd.Series(yf_target)
skewness=str(series.skew())
plt.legend([skewness],title=("skewness"),fontsize=20)
#plt.show()

plt.subplot(5,4,20)
stats.probplot(log_target,dist="norm",plot=plt)
plt.grid()
```

It's better to apply all the transformations and check which Transformation will give better result.</p>
