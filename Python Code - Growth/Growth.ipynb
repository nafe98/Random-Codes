```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.ticker import MultipleLocator

# ============================================================
# Data reconstructed from the original figure
# Unit: TWh/year
# ============================================================

years = np.arange(2010, 2031)

# Optimistic projection data
optimistic_consumer = np.array([
    380, 380, 370, 360, 350, 345, 340, 335, 335, 335,
    340, 345, 350, 370, 420, 480, 550, 620, 720, 900, 1100
])

optimistic_networking = np.array([
    550, 560, 560, 550, 540, 530, 520, 520, 515, 510,
    505, 500, 520, 560, 600, 630, 650, 680, 720, 780, 850
])

optimistic_datacenter = np.array([
    350, 350, 360, 350, 340, 340, 330, 330, 330, 330,
    330, 340, 430, 500, 560, 650, 750, 850, 1000, 1150, 1250
])

# Expected projection data
expected_consumer = np.array([
    700, 700, 700, 700, 700, 710, 720, 720, 730, 740,
    760, 780, 780, 760, 740, 720, 700, 680, 660, 650, 650
])

expected_networking = np.array([
    650, 660, 680, 700, 700, 700, 720, 740, 760, 780,
    800, 850, 1000, 1250, 1500, 1700, 2000, 2400, 3000, 3500, 3800
])

expected_datacenter = np.array([
    350, 380, 420, 470, 520, 580, 650, 750, 850, 950,
    1050, 1200, 1400, 1550, 1700, 1900, 2200, 2600, 3000, 3200, 3000
])

# ============================================================
# Figure styling
# ============================================================

plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman", "Times", "DejaVu Serif"],
    "font.size": 16,
    "axes.labelsize": 17,
    "axes.titlesize": 20,
    "axes.titleweight": "bold",
    "xtick.labelsize": 10,   # X-axis year font left as original
    "ytick.labelsize": 15,
    "legend.fontsize": 15,
    "axes.linewidth": 1.1,
    "figure.dpi": 150,
    "savefig.dpi": 900
})

colors = {
    "Consumer devices": "#B8B8B8",
    "Networking": "#2A9D8F",
    "Datacenters": "#7B2CBF"
}

# ============================================================
# Create figure
# ============================================================

fig, axes = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(14, 7.5),
    sharex=False
)

fig.patch.set_alpha(0.0)

# ============================================================
# Helper function
# ============================================================

def plot_ict_projection(
    ax,
    consumer,
    networking,
    datacenter,
    title,
    share_text,
    box_height
):
    ax.set_facecolor("none")

    ax.stackplot(
        years,
        consumer,
        networking,
        datacenter,
        labels=["Consumer devices", "Networking", "Datacenters"],
        colors=[
            colors["Consumer devices"],
            colors["Networking"],
            colors["Datacenters"]
        ],
        alpha=0.92
    )

    total = consumer + networking + datacenter

    ax.plot(
        years,
        total,
        color="black",
        linewidth=1.25,
        alpha=0.85
    )

    ax.set_xlim(2009.5, 2030.7)
    ax.set_ylim(0, 8000)

    ax.set_title(title, loc="left", pad=10)
    ax.set_ylabel("Annual energy footprint (TWh)")
    ax.set_xlabel("Year")

    ax.set_xticks(years)
    ax.set_xticklabels(
        years,
        rotation=0,
        ha="center",
        fontfamily="Times New Roman",
        fontsize=10
    )

    ax.tick_params(axis="x", pad=6)
    ax.tick_params(axis="both", width=1.1)

    ax.yaxis.set_major_locator(MultipleLocator(2000))

    ax.grid(
        True,
        which="major",
        axis="both",
        linestyle="--",
        linewidth=0.6,
        alpha=0.40
    )

    ax.add_patch(
        Rectangle(
            (2029.15, 0),
            1.35,
            box_height,
            fill=False,
            edgecolor="black",
            linewidth=1.5,
            zorder=5
        )
    )

    ax.text(
        2029.85,
        box_height + 180,
        "2030",
        ha="center",
        va="bottom",
        fontsize=14,
        fontweight="bold",
        fontfamily="Times New Roman"
    )

    ax.text(
        1.03,
        0.72,
        share_text,
        transform=ax.transAxes,
        ha="left",
        va="center",
        fontsize=16,
        fontfamily="Times New Roman",
        clip_on=False,
        bbox=dict(
            boxstyle="round,pad=0.35",
            facecolor="white",
            edgecolor="black",
            linewidth=0.8
        )
    )

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

# ============================================================
# Top plot
# ============================================================

plot_ict_projection(
    ax=axes[0],
    consumer=optimistic_consumer,
    networking=optimistic_networking,
    datacenter=optimistic_datacenter,
    title="Optimistic ICT Energy Projection",
    share_text="2030 estimate:\n7% of global\nelectricity demand",
    box_height=3450
)

# ============================================================
# Bottom plot
# ============================================================

plot_ict_projection(
    ax=axes[1],
    consumer=expected_consumer,
    networking=expected_networking,
    datacenter=expected_datacenter,
    title="Expected ICT Energy Projection",
    share_text="2030 estimate:\n20% of global\nelectricity demand",
    box_height=7800
)

# ============================================================
# Shared legend
# ============================================================

handles, labels = axes[0].get_legend_handles_labels()

fig.legend(
    handles,
    labels,
    loc="upper center",
    ncol=3,
    frameon=False,
    bbox_to_anchor=(0.47, 0.995)
)

plt.tight_layout(rect=[0.04, 0.05, 0.82, 0.95])

# ============================================================
# Save outputs
# ============================================================

plt.savefig(
    "ict_energy_projection_final_wide_HD_transparent.png",
    dpi=900,
    bbox_inches="tight",
    pad_inches=0.08,
    transparent=True
)

plt.savefig(
    "ict_energy_projection_final_wide_transparent.pdf",
    bbox_inches="tight",
    pad_inches=0.08,
    transparent=True
)

plt.savefig(
    "ict_energy_projection_final_wide_transparent.svg",
    bbox_inches="tight",
    pad_inches=0.08,
    transparent=True
)

plt.show()
```
