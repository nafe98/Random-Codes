```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a DataFrame with the given data
data = {
    'Category': ['Q1', 'Q2', 'Q3', 'Not indexed'],
    'Count': [15, 7, 3, 4]
}
df = pd.DataFrame(data)

# Set the aesthetic style for the plot
sns.set_style("whitegrid")

# Create a bar plot
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Category', y='Count', data=df, palette='viridis')

# Add titles and labels
plt.title('Number of Articles by SJR Journal Ranking Category', fontsize=16)
plt.xlabel('Journal Ranking Category', fontsize=14)
plt.ylabel('Number of Articles', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Annotate each bar with its value
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center',
                     xytext=(0, 10),
                     textcoords='offset points',
                     fontsize=12, color='black')

# Adjust layout to fit all elements
plt.tight_layout()

# Save the bar plot to a file
plt.savefig('bar_plot.png')

# Display the plot
plt.show()
```
